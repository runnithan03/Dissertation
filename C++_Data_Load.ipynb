```python
%%writefile load_data.cpp

#include <iostream>
#include <fstream>
#include <vector>
#include <string>
#include <sstream>
#include <iomanip>

using namespace std;

// Function to split a line by a delimiter
vector<string> split(const string& line, char delimiter) {
    vector<string> tokens;
    stringstream ss(line);
    string token;
    while (getline(ss, token, delimiter)) {
        tokens.push_back(token);
    }
    return tokens;
}

// Function to read the CSV file and store data in a 2D vector
vector<vector<string>> readCSV(const string& filename) {
    vector<vector<string>> data;
    ifstream file(filename);

    if (!file.is_open()) {
        cerr << "Error: Could not open file '" << filename << "'." << endl;
        return data;
    } else {
        cout << "File '" << filename << "' opened successfully!" << endl;
    }

    string line;
    int lineCount = 0;
    while (getline(file, line)) {
        vector<string> row = split(line, ',');
        data.push_back(row);
        lineCount++;

        // Stop after loading 10 rows for debugging
        if (lineCount <= 10) {
            cout << "Row " << lineCount << ": ";
            for (const auto& cell : row) {
                cout << cell << " ";
            }
            cout << endl;
        }
    }

    cout << "Total rows loaded: " << lineCount << endl;
    file.close();
    return data;
}

int main() {
    string filename = "Profit ESG dataset.csv";  // Use the exact filename

    // Read the CSV data
    vector<vector<string>> data = readCSV(filename);

    // Check if data was successfully loaded
    if (data.empty()) {
        cout << "No data loaded. Please ensure that 'Profit ESG dataset.csv' is available and formatted correctly." << endl;
        return 1;
    }

    return 0;
}
```
